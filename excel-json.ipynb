## Preparação

```python
import os
import math

import glob
import pandas as pd

from openpyxl import load_workbook
```

```python
os.makedirs("files", exist_ok=True)
os.makedirs("output/csv", exist_ok=True)
os.makedirs("output/json", exist_ok=True)
```

```python
paths = glob.glob("files/*.xlsx")
```

## Leitura dos arquivos XLSX (transforma em pd.dataframe)

```python
dataframes = {}

for i, path in enumerate(paths, start=1):
    try:
        df = pd.read_excel(path)
    except Exception as e:
        print("Erro ao ler", path, ":", e)
        continue

    nome = os.path.splitext(os.path.basename(path))[0]

    dataframes[nome] = df
    print(nome, "carregado:", path)
```

## Conversão para CSV com no máximo 20k linhas

```python
limite_de_linhas = 20000

for nome, df in dataframes.items():
    total_de_linhas = len(df)
    total_de_arquivos = math.ceil(total_de_linhas / limite_de_linhas)
    for i in range(total_de_arquivos):
        inicio = i * limite_de_linhas
        fim = inicio + limite_de_linhas
        selecao = df.iloc[inicio:fim]
        
        selecao.to_csv(f"output/csv/{nome}(pt{i+1}).csv", index=False)
        print(f"Arquivo {nome}_parte_{i+1} com as linhas de {inicio} a {fim}, gerado com sucesso!")
```

## Conversão para JSON

```python
for nome, df in dataframes.items():
    df.to_json(f"output/json/{nome}.json", orient="records")
    print(f"Arquivo {nome}.json gerado com sucesso!")
```
